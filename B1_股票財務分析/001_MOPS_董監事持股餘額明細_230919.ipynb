```python
#董監事持股餘額明細 #pd.read_html()

import requests
import pandas as pd

def get_董監事持股餘額明細(co_id,year,month): #定義函數

    url='https://mops.twse.com.tw/mops/web/ajax_stapap1' #抓資料頁面的網址
    parameter = { 'firstin' : '1', #電腦要送出的參數
                    'co_id' : str(co_id),
                    'year'  : str(year),
                    'month' : str(month), }
    res = requests.post(url, data = parameter) #抓到的資料存到res
    df = pd.read_html(res.text)[3] #把需要的資料存在df

    return df

get_董監事持股餘額明細(2330,111,8) #使用函數，把(股票代號,年份,日期)資料抓下來
```
