### 資料分析

```python
#資料類型轉換
df_unclean.drop('index', axis=1, inplace=True)
```

```python
df_unclean
```

```python
# 拆分 'Company Name' 欄位為 'Company' 和 'company_rating'
df_unclean['company_name'] = df_unclean['Company Name'].apply(lambda x: x.split('\n')[0])
df_unclean['company_rating'] = df_unclean['Company Name'].apply(lambda x: float(x.split('\n')[1]) if len(x.split('\n')) > 1 else None)

# 將原 'Rating' 欄位名稱更改為 'job_rating'
df_unclean.rename(columns={'Rating': 'job_rating'}, inplace=True)

# 顯示更新後的資料的前5行，確認新欄位是否正確
df_unclean.head()
```

```python
df_unclean.drop('Company Name', axis=1, inplace=True)
```

```python
unique_name = df_unclean['company_name'].unique()
unique_name, len(unique_name)
```

```python
#Job Title
unique_job = df_unclean['Job Title'].unique()
unique_job, len(unique_job)
```

```python
#分組
# 定義一個函數來分組 job_title
def categorize_job_title(title):
    title = title.lower()
    if 'data scientist' in title or 'scientist' in title:
        return 'Data Scientist Group'
    elif 'data analyst' in title:
        return 'Data Analyst Group'
    elif 'data engineer' in title:
        return 'Data Engineer Group'
    elif 'engineer' in title:
        return 'Other Engineers Group'
    else:
        return 'Others Group'

# 應用函數來創建一個新的欄位
df_unclean['job_title_group'] = df_unclean['Job Title'].apply(categorize_job_title)

# 檢查分組的結果
df_unclean['job_title_group'].value_counts()
```

```python
df_unclean.head()
```

```python
#第三個欄位：Salary Estimate->分別拆成最高、最低薪水欄位

# 檢查 'Salary Estimate' 欄位的唯一值
df_unclean['Salary Estimate'].unique()
```

```python
# 使用replace函數將'(Employer est.)'修改為' (Employer est.)'，使其與其他值的格式一致
df_unclean['Salary Estimate'] = df_unclean['Salary Estimate'].replace('$145K-$225K(Employer est.)', '$145K-$225K (Employer est.)')
df_unclean['Salary Estimate'].unique()
```

```python
#將 Salary Estimate 欄位拆分為 min_salary、max_salary 和 avg_salary 三個新欄位

# 拆分 'Salary Estimate' 欄位為 'min_salary' 和 'max_salary'
df_unclean['min_salary'] = df_unclean['Salary Estimate'].apply(lambda x: int(x.split('-')[0].replace('$', '').replace('K', '')))
df_unclean['max_salary'] = df_unclean['Salary Estimate'].apply(lambda x: int(x.split('-')[1].replace('$', '').replace('K', '').split(' ')[0]))

# 計算平均薪資
df_unclean['avg_salary'] = (df_unclean['min_salary'] + df_unclean['max_salary']) / 2

# 顯示更新後的資料的前5行，確認新欄位是否正確
df_unclean.head()
```

```python
#拆分出城市與州欄位
# 拆分 'Location' 欄位為 'City' 和 'State'
df_unclean['City'] = df_unclean['Location'].apply(lambda x: x.split(',')[0].strip())
df_unclean['State'] = df_unclean['Location'].apply(lambda x: x.split(',')[1].strip() if len(x.split(',')) > 1 else None)

# 顯示更新後的資料的前5行，確認新欄位是否正確
df_unclean.head()
```

```python
# 刪除 'Size' 欄位中值為 '-1' 的行 #27筆
df_unclean = df_unclean[df_unclean['Size'] != '-1']

# 顯示更新後的資料的前5行，確認資料是否已被刪除
df_unclean.info()
```

```python
# 計算公司年齡並新增到 'company_age' 欄位
current_year = 2023  # 當前年份
df_unclean['company_age'] = current_year - df_unclean['Founded']

# 顯示更新後的資料的前5行，確認新欄位是否正確
df_unclean.head()
```

```python
# 檢查 'Industry' 欄位的唯一值
unique_industries = df_unclean['Industry'].unique()
unique_industries, len(unique_industries)
```

```python
# 檢查 'Sector' 欄位的唯一值
unique_sectors = df_unclean['Sector'].unique()
unique_sectors, len(unique_sectors)
```

```python
# 對 'Type of ownership' 欄位的唯一值進行編碼，從 1 開始
ownership_mapping = {ownership: idx+1 for idx, ownership in enumerate(df_unclean['Type of ownership'].unique())}

# 新增 'Type_of_ownership_gp' 欄位，用於存儲所有權類型的分組
df_unclean['Type_of_ownership_gp'] = df_unclean['Type of ownership'].map(ownership_mapping)

# 顯示更新後的資料的前5行，確認新欄位是否正確
df_unclean.head()
```

```python
# 檢查 'Revenue' 欄位的唯一值
unique_revenues = df_unclean['Revenue'].unique()
unique_revenues, len(unique_revenues)
```

```python
# 計算 'Revenue' 欄位中值為 'Unknown / Non-Applicable' 的行數
count_unknown_revenue = df_unclean[df_unclean['Revenue'] == 'Unknown / Non-Applicable'].shape[0]
count_unknown_revenue
```

```python
# 將 'Competitors' 欄位中的字符串拆分成列表
df_unclean['Competitors'] = df_unclean['Competitors'].apply(lambda x: x.split(',') if x != '-1' else [])

# 顯示更新後的資料的前5行，確認拆分是否成功
df_unclean.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#job tilte
# 設置圖形大小和風格
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

# 繪製 job_title_group 的條形圖
sns.countplot(y='job_title_group', data=df_unclean, order=df_unclean['job_title_group'].value_counts().index)

plt.title('Distribution of job_title_group') #圖片標題
plt.xlabel('Count') #x軸
plt.ylabel('Job Title Group') #左軸

plt.show()
```

```python
# Type_of_ownership_gp
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

# 繪製 Type_of_ownership_gp 的條形圖
sns.countplot(y='Type of ownership', data=df_unclean, order=df_unclean['Type of ownership'].value_counts().index)

plt.title('Distribution of Type of ownership')
plt.xlabel('Count')
plt.ylabel('Type of ownership')

plt.show()
```
