#### 將取資料寫成函數 Data.py

```python
# 將取資料寫成函數 Data.py
import yfinance as yf
import pandas as pd
import os

#定義變數
#prod='0050.TW'
#starday='2022-01-1'
#endday='2022-07-15'

def getDataYF(prod,starday,endday):
    #定義存檔名稱
    backup_file='data\\YF_%s_%s_%s.csv'%(prod,starday,endday)

    #第一步:确認是否有本地端檔案
    if not os.path.exists(backup_file):
        #如果沒有我就跟之後寫
        data=yf.download(prod,start = starday, end =endday)
        data.to_csv(backup_file)
        print(f'新增{backup_file}檔案')

    #如果有我就用本地端檔案即可
    else:
        data=pd.read_csv(backup_file)
        #檔案取得後，須將時間改成索引
        data=data.set_index('Date')
        print(f'本地端已有{backup_file}檔案')
    return data

#df=getDataYF(prod,starday,endday)
```
