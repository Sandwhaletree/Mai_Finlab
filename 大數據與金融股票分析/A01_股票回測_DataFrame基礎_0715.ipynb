```python
import pandas as pd
# 定義list
a=[ [1,2,3],[2,3,4],[3,4,5] ]
# 從list 轉換成dataframe 的方法
pd.DataFrame(a)
```

```python
# 定義dictionary
b={    'open':[1,2,3],
       'high':[7,8,9],
       'close' :[3,4,5]}
# 從dictionary 轉換為dataframe 的方法
pd.DataFrame(b)
```

```python
# 組成函示庫
# function
# class
# 1- 屬性 (variable)
# 2- 方法 (function)
```

```python
# 查看dataframe 屬性
a=pd.DataFrame(b)
a.ndim # 維度
a.shape # 面積
a.dtypes # 資料型態
```

```python
# 取得行數
a.shape[0]
```

```python
# 如何呼叫欄位
a['open'] #老師習慣這種方法，因為有時候欄位是中文，用a.open感覺怪怪的
a.open
```

```python
# 針對指定行數 去新增欄位
a['low']=[0,1,2] #資料筆數與原資料是一樣的
a
```

```python
# 直接將新欄位調整成統一值
a['volume']= 100
a
```

#### loc (locate)

```python
# 如何取用資料-loc # locate
# a.loc[行,列] # 用欄位名稱來取得指定資料的方法
a.loc[1,'low']
```

```python
#牛刀小試：抓到2的行 close的值
a.loc[2,'close']
```

```python
# 抓到1,2兩行
a.loc[1:2,'open']
```

```python
# 抓到值為1以後的行的值
a.loc[1:,'open']
```

```python
#抓幾行的最高價
max(a.loc[1:2,'high'])
```

#### iloc(index locate)

```python
# 如何取用資料-iloc # index locate
# 是用欄位索引來取得指定資料的方法
a.iloc[2,3] # 實際上取得low的值
```

```python
# df的行索引
a.index
```

```python
# df的欄位索引
a.columns
```

```python
#列表推導式
[ i for i in a.index ]
```

```python
#tmplist=[1,2,3]
#[ i for i in tmplist ]
#  ^     ^       ^
#運算式
#      迴圈變數 目標的集合

#[ 1, 2, 3 ] <- 集合
#(元素)
[ i for i in range(1,10) ]
```

```python
# dataframe 篩選資料
a[ a['high'] > 7 ]
```

```python
# 取得 not 的條件判斷
a[ ~(a['high'] > 7) ]
```

```python
# 複合判斷式
a[ (a['high'] > 7)&(a['open'] > 2) ]
```
