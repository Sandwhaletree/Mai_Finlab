# 列表
清單是社群的資料容器，可以包含不同類型的元素。它們用方括號[]定義，並使用分隔,元素。

```python
# 建立列表
my_list = [1, 2, 3, 4, 5]
my_list
```

```python
# 取得單個元素
first_element = my_list[0]
print(first_element)
last_element = my_list[-1]
print(last_element)
```

```python
# 取得多個元素
sliced_list = my_list[1:4]
sliced_list
```

```python
# 附加元素
my_list.append(6)
print(my_list)
```

```python
# 擷取元素
get_one = my_list.pop(0)
print(get_one)
print(my_list)
```

# 字典

字典是資料結構的一個鍵值，用花括號{}定義。每個鍵都與一個值關聯。

```python
# 建立字典
my_dict = {"name": "Alice", "age": 30, "city": "New York"}
my_dict
```

```python
# 查看值
name = my_dict["name"]
name
```

```python
# 新增鍵值
my_dict["job"] = "Engineer"
my_dict
```

# 集合（Set）

集合是無序且不允許重複元素的容器，用花的{}定義。

```python
# 建立集合
my_set = {1, 2, 3, 4, 5}
```

```python
# 新增元素
my_set.add(6)
```

```python
# 聯集運算
union_set = my_set.union({4, 5, 6})
print(union_set)
```

```python
# 交集運算
intersection_set = my_set.intersection({3, 4, 5})
print(intersection_set)
```

# 程式結構 - 條件判斷
條件語句用於根據不同的條件執行不同的操作。常見的條件語句包括if、elif（else if）和else

```python
x = 10
if x > 5:
    print("x > 5")
elif x == 5:
    print("x == 5")
else:
    print("x < 5")
```

# 程式結構 - 迴圈
循環用於重複執行一組操作，有兩種常見類型的循環：for循環和while循環

```python
# for 案例
for i in range(5):
    print(i)
```

```python
# while 案例
count = 0
while count < 5:
    print(count)
    count += 1
```

# 函數（Function）

函數是一個可重複使用的程式碼區塊，它可以接受參數並傳回一個值。函數定義以關鍵字開始，後面跟著函數名稱def、參數列表和冒號。函數體縮排並包含函數的操作。

```python
# 函数示例
def add_numbers(x, y):
    result = x + y
    return result

sum_result = add_numbers(3, 5)
print(sum_result)
```

# 異常處理（Exception Handling）
異常處理可讓您處理可能出現的錯誤情況，以確保程式不會崩潰。常見的異常處理結構包括try、except和finally

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("錯誤發生")
finally:
    print("無論正常或錯誤都會執行的區塊")
```
