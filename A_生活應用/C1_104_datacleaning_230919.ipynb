```python
# Colab 進行matplotlib繪圖時顯示繁體中文
# 下載台北思源黑體並命名taipei_sans_tc_beta.ttf，移至指定路徑
!wget -O TaipeiSansTCBeta-Regular.ttf https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download

import matplotlib as mpl
import matplotlib.pyplot as plt #導入Matplotlib子模塊pyplot(函數繪製圖表)
from matplotlib.font_manager import fontManager

# 改style要在改font之前
# plt.style.use('seaborn')

fontManager.addfont('TaipeiSansTCBeta-Regular.ttf')
mpl.rc('font', family='Taipei Sans TC Beta')
```

```python
#哪個城市中工作數最多
#畫長條圖 #df.groupby('x軸')['y軸'].count().plot(kind='bar', title="標題")

# 繪製圖表
job_merge.groupby("城市")["工作數"].count().plot(kind='bar', title="哪個城市中工作數最多")
```

```python
#長條圖(有排序)
job_merge['城市'].value_counts().plot(kind="bar")
```

```python
# 使用pandas的boxplot方法來繪制箱形圖
job_merge.boxplot(column='薪資', by='城市', figsize=(10, 6), vert=True) #vert=False 水平箱型圖；(預設)Ture=垂直箱型圖
# 設置圖表的標題和x、y軸的標籤
plt.title('各城市薪資的箱形圖')
plt.xlabel('城市')
plt.ylabel('薪資')

# 顯示圖表
plt.show()
```

```python
# 分組城市之箱型圖繪畫

# 使用groupby方法對數據分組
job_group = job_merge.groupby("城市")

# 定義一個空列表，用於存儲每個城市的薪資數據
salaries = []

# 定義一個空列表，用於存儲每個城市的名稱
cities = []

# 遍歷每個城市
for city, data in job_group:
    # 將薪資數據存儲到列表中
    salaries.append(data['薪資'])

    # 將城市名稱存儲到列表中
    cities.append(city)

# 繪制箱形圖
plt.figure(figsize=(10, 6))  # 指定圖表的大小
plt.boxplot(salaries, labels=cities, vert=False)

# 設置圖表的標題和x、y軸的標籤
plt.title('不同城市薪資的箱形圖')
plt.xlabel('城市')
plt.ylabel('薪資')

# 顯示圖表
plt.show()
```
