```python
import requests
from bs4 import BeautifulSoup

# 目標網址
url = 'https://www.lotto-8.com/listltobigbbk.asp'

# 發送HTTP請求
r1 = requests.get(url)
r1.encoding="big-5"

# 使用BeautifulSoup來解析HTML
soup = BeautifulSoup(r1.text, 'lxml')

# 找到表格
table = soup.find('table', {'class': 'auto-style4'})

# 初始化一個空列表來儲存解析後的數據
parsed_data = []

# 遍歷表格的每一行
for row in table.find_all('tr')[1:]:  # 跳過表頭
    cells = row.find_all('td')

    # 獲取日期
    date = cells[0].text.strip().replace('\n', ' ')

    # 獲取大樂透中獎號碼
    numbers = cells[1].text.strip().replace('\xa0', ' ')

    # 獲取特別號
    special_number = cells[2].text.strip()

    parsed_data.append({
        'Date': date,
        'Numbers': numbers,
        'Special Number': special_number
    })

# 輸出解析後的數據
for data in parsed_data:
    print(f"Date: {data['Date']}, Numbers: {data['Numbers']}, Special Number: {data['Special Number']}")

print(len(parsed_data))
```

```python
import requests
from bs4 import BeautifulSoup

# 初始化一個空列表來儲存解析後的數據
parsed_data = []

# 遍歷1到10頁
for page in range(1, 30):
    # 目標網址
    url = f'https://www.lotto-8.com/listltobigbbk.asp?indexpage={page}&orderby=new'

    # 發送HTTP請求
    r = requests.get(url)
    r.encoding = "big-5"

    # 使用BeautifulSoup來解析HTML
    soup = BeautifulSoup(r.text, 'lxml')

    # 找到表格
    table = soup.find('table', {'class': 'auto-style4'})

    # 遍歷表格的每一行
    for row in table.find_all('tr')[1:]:  # 跳過表頭
        cells = row.find_all('td')

        # 獲取日期
        date = cells[0].text.strip().replace('\n', ' ')

        # 獲取大樂透中獎號碼
        numbers = cells[1].text.strip().replace('\xa0', ' ')

        # 獲取特別號
        special_number = cells[2].text.strip()

        parsed_data.append({
            'Date': date,
            'Numbers': numbers,
            'Special Number': special_number
        })

# 輸出解析後的數據
for data in parsed_data:
    print(f"Date: {data['Date']}, Numbers: {data['Numbers']}, Special Number: {data['Special Number']}")
```

```python
# 寫入CSV文件
import csv
with open('lotto0914.csv', 'w', newline='') as csvfile:
    fieldnames = ['Date', 'Numbers', 'Special Number']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for data in parsed_data:
        writer.writerow(data)

print("CSV檔案完成")
```

### 中秋節加碼

```python
import requests
from bs4 import BeautifulSoup
import csv

# 目標網址
url = 'https://www.pilio.idv.tw/ltobig/listspecial_100.asp'

# 發送HTTP請求
r = requests.get(url)
r.encoding = "big-5"

# 使用BeautifulSoup來解析HTML
soup = BeautifulSoup(r.text, 'lxml')

# 找到表格
table = soup.find('table', {'style': 'width:100%'})

# 初始化一個空列表來儲存解析後的數據
parsed_data = []

# 遍歷表格的每一行（跳過第一行標題）
for row in table.find_all('tr')[1:]:
    cell = row.find('td')

    # 分割日期和組別
    date_group = cell.text.strip().split("___")
    date = date_group[0]
    group = date_group[1]

    # 獲取中獎號碼
    numbers = cell.find('span').text.strip()

    parsed_data.append({
        'Date': date,
        'Group': group,
        'Numbers': numbers
    })

# 寫入CSV文件
with open('moonlotto0914.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Date', 'Group', 'Numbers']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for data in parsed_data:
        writer.writerow(data)

print("CSV file has been created.")
```
