```python
import pandas as pd
import numpy as np
os.chdir('./drive/MyDrive/cloud_file/kaggle/supermarket-sales')
sales = pd.read_csv('./supermarket_sales - Sheet1.csv')
sales
```

#### 👉**1.欄位項目個數** ####
* .value_counts()
* groupby('col').size()

```python
#檢查Branch欄位中，有多少筆不重複資料
#方法一
sales['Branch'].value_counts()
```

```python
#方法二
sales.groupby('Branch').size()
```

#### 👉**2.欄位分組運算** ####

```python
sales.groupby('Branch').sum()
```

#### 👉**3.多欄位分組運算** ####
* 假設我們想知道這家超市的三間分店中，來自會員與非會員的平均銷售額分別是多少，該怎麼做呢？

```python
sales.groupby(['Branch', 'Customer type'])[['Total']].mean()
```

#### 👉**4.取得欄位分組資料** ####

```python
# 取出 Branch 欄位中，顯示為 A 的資料，共 340 筆
group_A = sales.groupby('Branch').get_group('A')
group_A
```
