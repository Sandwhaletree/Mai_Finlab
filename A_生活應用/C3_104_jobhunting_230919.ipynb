```python
#å°‡å„å€‹è·ç¼ºå…§å®¹å­˜åˆ°csvæª”
import requests
import bs4
import csv

title_job = str(input("è«‹è¼¸å…¥æ¬²æŸ¥è©¢ä¹‹è·ç¼ºå·¥ä½œ")) #æœå°‹é—œéµå­—
print('æº–å‚™æœå°‹title:',title_job)

url = f"https://www.104.com.tw/jobs/search/?keyword={title_job}&order=1&jobsource=2018indexpoc&ro=0"
htmlFile = requests.get(url)
ObjSoup=bs4.BeautifulSoup(htmlFile.text, 'lxml')
jobs = ObjSoup.find_all('article',class_='js-job-item') #æœå°‹æ‰€æœ‰è·ç¼º

import datetime
date = str(datetime.date.today())
fn= date+'_104è·ç¼º_'+ title_job +'.csv' #å–CSVæª”å

columns_name=['è·ä½', 'å…¬å¸','è–ªè³‡', 'åœ°å€','ç¶²å€'] #ç¬¬ä¸€æ¬„çš„åç¨±
with open(fn,'w',newline='',) as csvFile: ##å®šç¾©csvçš„å¯«å…¥æª”,ä¸¦ä¸”æ¯æ¬¡å¯«å…¥å®Œæœƒæ›ä¸‹ä¸€è¡Œ
    writer = csv.DictWriter(csvFile, fieldnames=columns_name) #å®šç¾©å¯«å…¥å™¨
    writer.writeheader()

    for job in jobs:
      job_name=job.find('a',class_="js-job-link").text #è·ç¼º
      job_company=job.get('data-cust-name') #å…¬å¸åç¨±
      job_1oc=job.find('ul', class_='job-list-intro').find('li').text  #åœ°å€
      try:
        job_pay=job.find('a',class_='b-tag--default').text  #è–ªè³‡
      except:
        job_pay="é¢è­°"
      job_url=job.find('a').get('href') #ç¶²å€

      writer.writerow({'è·ä½':job_name ,'å…¬å¸':job_company,'è–ªè³‡':job_pay,'åœ°å€':job_1oc,'ç¶²å€':job_url})

print('æª”æ¡ˆè¼¸å‡ºå®Œæˆ:'+fn)
```

### ğŸ‘‰**3.çˆ¬å–å¤šé 104è³‡æ–™ï¼Œä¸¦å­˜æˆcsvæª”**###
* è¼¸å…¥æ¬²æŸ¥è©¢ä¹‹è·ç¼ºå·¥ä½œ
* å°‡è¼¸å‡º104ç¶²ç«™"50é "ä¹‹æŸ¥æ‰¾å·¥ä½œä¹‹è¨Šï¼Œä¸¦è¼¸å‡ºç‚ºcsvæª”

```python
#çˆ¬å–5é çš„è·ç¼ºå…§å®¹,å°‡å„å€‹è·ç¼ºå…§å®¹å­˜åˆ°csvæª”
import requests
import bs4
import csv
import random,time

title_job = str(input("è«‹è¼¸å…¥æ¬²æŸ¥è©¢ä¹‹è·ç¼ºå·¥ä½œ")) #æœå°‹é—œéµå­— #å¤šé—œéµå­—æœå°‹ å¦‚:Business Analystã€Data analystã€Python
print('æº–å‚™æœå°‹title:',title_job)

url_A=f"https://www.104.com.tw/jobs/search/?ro=o&kwop=7&keyword={title_job}&order=15&asc=0&page="
url_B = '&mode=s&jobsource=2018indexpoc'

all_job_datas=[ ]
for page in range(1,50+1): #ä¸€é 20å€‹è·ç¼ºï¼›5é 100å€‹è·ç¼º
  url = url_A+str(page)+url_B
  print(url)
  htmlFile = requests.get(url)
  ObjSoup=bs4.BeautifulSoup(htmlFile.text,'lxml')
  jobs = ObjSoup.find_all('article',class_='js-job-item') #æœå°‹æ‰€æœ‰è·ç¼º

  for job in jobs:
      job_name=job.find('a',class_="js-job-link").text #è·ç¼º
      job_company=job.get('data-cust-name') #å…¬å¸åç¨±
      job_1oc=job.find('ul', class_='job-list-intro').find('li').text  #åœ°å€
      try:
        job_pay=job.find('a',class_='b-tag--default').text  #è–ªè³‡
      except:
        job_pay="é¢è­°"
      job_url=job.find('a').get('href') #ç¶²å€
      job_data={'è·ä½':job_name ,'å…¬å¸':job_company,'åœ°å€':job_1oc,'è–ªè³‡':job_pay,'ç¶²å€':job_url}
      all_job_datas.append(job_data)
  time.sleep(random.randint(1,3))

import datetime
date = str(datetime.date.today())
fn= date+'_104è·ç¼ºå¤šé _'+ title_job +'.csv' #å–CSVæª”å

columns_name=['è·ä½', 'å…¬å¸', 'åœ°å€', 'è–ªè³‡','ç¶²å€']
with open(fn,'w',newline='',) as csvFile: #å®šç¾©csvçš„å¯«å…¥æª”,ä¸¦ä¸”æ¯æ¬¡å¯«å…¥å®Œæœƒæ›ä¸‹ä¸€è¡Œ
    writer = csv.DictWriter(csvFile, fieldnames=columns_name) #å®šç¾©å¯«å…¥å™¨
    writer.writeheader()
    for data in all_job_datas:
        writer.writerow(data)

print('æª”æ¡ˆè¼¸å‡ºå®Œæˆ:'+fn)
```
