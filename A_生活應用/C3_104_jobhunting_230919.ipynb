### 👉**2.爬取104單頁資料，並存成csv檔**###
* 輸入欲查詢之職缺工作
* 將輸出104網站單頁之查找工作之訊，並輸出為csv檔

```python
#將各個職缺內容存到csv檔
import requests
import bs4
import csv

title_job = str(input("請輸入欲查詢之職缺工作")) #搜尋關鍵字
print('準備搜尋title:',title_job)

url = f"https://www.104.com.tw/jobs/search/?keyword={title_job}&order=1&jobsource=2018indexpoc&ro=0"
htmlFile = requests.get(url)
ObjSoup=bs4.BeautifulSoup(htmlFile.text, 'lxml')
jobs = ObjSoup.find_all('article',class_='js-job-item') #搜尋所有職缺

import datetime
date = str(datetime.date.today())
fn= date+'_104職缺_'+ title_job +'.csv' #取CSV檔名

columns_name=['職位', '公司','薪資', '地址','網址'] #第一欄的名稱
with open(fn,'w',newline='',) as csvFile: ##定義csv的寫入檔,並且每次寫入完會換下一行
    writer = csv.DictWriter(csvFile, fieldnames=columns_name) #定義寫入器
    writer.writeheader()

    for job in jobs:
      job_name=job.find('a',class_="js-job-link").text #職缺
      job_company=job.get('data-cust-name') #公司名稱
      job_1oc=job.find('ul', class_='job-list-intro').find('li').text  #地址
      try:
        job_pay=job.find('a',class_='b-tag--default').text  #薪資
      except:
        job_pay="面議"
      job_url=job.find('a').get('href') #網址

      writer.writerow({'職位':job_name ,'公司':job_company,'薪資':job_pay,'地址':job_1oc,'網址':job_url})

print('檔案輸出完成:'+fn)
```

### 👉**3.爬取多頁104資料，並存成csv檔**###
* 輸入欲查詢之職缺工作
* 將輸出104網站"50頁"之查找工作之訊，並輸出為csv檔

```python
#爬取5頁的職缺內容,將各個職缺內容存到csv檔
import requests
import bs4
import csv
import random,time

title_job = str(input("請輸入欲查詢之職缺工作")) #搜尋關鍵字 #多關鍵字搜尋 如:Business Analyst、Data analyst、Python
print('準備搜尋title:',title_job)

url_A=f"https://www.104.com.tw/jobs/search/?ro=o&kwop=7&keyword={title_job}&order=15&asc=0&page="
url_B = '&mode=s&jobsource=2018indexpoc'

all_job_datas=[ ]
for page in range(1,50+1): #一頁20個職缺；5頁100個職缺
  url = url_A+str(page)+url_B
  print(url)
  htmlFile = requests.get(url)
  ObjSoup=bs4.BeautifulSoup(htmlFile.text,'lxml')
  jobs = ObjSoup.find_all('article',class_='js-job-item') #搜尋所有職缺

  for job in jobs:
      job_name=job.find('a',class_="js-job-link").text #職缺
      job_company=job.get('data-cust-name') #公司名稱
      job_1oc=job.find('ul', class_='job-list-intro').find('li').text  #地址
      try:
        job_pay=job.find('a',class_='b-tag--default').text  #薪資
      except:
        job_pay="面議"
      job_url=job.find('a').get('href') #網址
      job_data={'職位':job_name ,'公司':job_company,'地址':job_1oc,'薪資':job_pay,'網址':job_url}
      all_job_datas.append(job_data)
  time.sleep(random.randint(1,3))

import datetime
date = str(datetime.date.today())
fn= date+'_104職缺多頁_'+ title_job +'.csv' #取CSV檔名

columns_name=['職位', '公司', '地址', '薪資','網址']
with open(fn,'w',newline='',) as csvFile: #定義csv的寫入檔,並且每次寫入完會換下一行
    writer = csv.DictWriter(csvFile, fieldnames=columns_name) #定義寫入器
    writer.writeheader()
    for data in all_job_datas:
        writer.writerow(data)

print('檔案輸出完成:'+fn)
```
