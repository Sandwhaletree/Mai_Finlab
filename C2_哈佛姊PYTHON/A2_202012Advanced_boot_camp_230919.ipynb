````python
import pandas as pd
file_name = './L2/week2/baby_names_ca.TXT'
columns = ['State', 'Sex', 'Year', 'Name', 'Count']
#讀取檔案
babynames =pd.read_csv(file_name,names=columns)
babynames
````

###### 👉 **1、尋找英文名(如jennifer)之熱門度變化** ######

````python
#設計熱門度變化最大程式指令
def ammd(series):
    return max(series) - min(series)
````

````python
#查看有jennifer之表格
babynames.query("Name == 'Jennifer'")
````

````python
#統計各年度jennifer之次數
jennifer_counts = babynames.query("Name == 'Jennifer'")["Count"]
jennifer_counts
````

````python
#jennifer出現最多次數
max(jennifer_counts)
````

````python
#jennifer出現最少次數
min(jennifer_counts)
````

````python
#jennifer最多-最少次數(熱門度變化最大)
ammd(jennifer_counts)
````

###### 👉 **2、查看各英文名之熱門度變化** ######

````python
#尋找英文名唯一值 #ANS:總共19645個名字
len(babynames["Name"].unique())
````

````python
#用空字典寫入英文名之熱門度變化
ammd_of_babyname_counts = {} #定義dict

for name in sorted(babynames["Name"].unique()[:10]): #sorted 僅是幫名字排順序 #名字從小到大排序過且唯一名之list
    counts_of_current_name = babynames[babynames["Name"] == name]["Count"] #抓出"Name"欄，alice各年度(所有名字)之Count數字 #series
    ammd_of_babyname_counts[name] = ammd(counts_of_current_name) #定義VALUE #將ammd(counts_of_current_name)值，存入dict[name]
````

````python
#查詢ammd_of_babyname_counts型態 #ANS:dict型態
type(ammd_of_babyname_counts)
````

````python
#把ammd_of_babyname_counts轉換成pandas Series
ammd_of_babyname_counts = pd.Series(ammd_of_babyname_counts)
````

````python
#查看各英文名之熱門度變化
ammd_of_babyname_counts
````

````python
#最熱門的名字共3716 #
ammd_of_babyname_counts.max()
````

````python
#尋找3716的名字
ammd_of_babyname_counts.sort_values(ascending=False)
````

````python
#Michael曾經很熱門、曾經很不熱門
babynames.groupby("Name").agg(ammd).sort_values(by="Count", ascending=False)
````

````python
#驗證 Michael曾經很熱門、曾經很不熱門
babynames[babynames['Name']=='Michael']['Count'].max()
````

````python
babynames[babynames['Name']=='Michael']['Count'].min()
````

````python
#每年度選出最熱門名字之次數
babynames.groupby("Year").agg(ammd)
````

````python
#最熱門-最冷門於1920年變化不大；但到1960年變化卻很大
babynames.groupby("Year").agg(ammd).plot()
````

````python
#標準化國家名
def canonicalize_county(county_name):
    return (
        county_name
        .lower()               # lower case
        .replace(' ', '')      # remove spaces
        .replace('&', 'and')   # replace &
        # your turn now - do the following
#         .replace('.', '') #remove(dot)
#         .replace('county', '') #remove county
#         .replace('parish', '') #remove parish
    )
````

````python
county_and_pop['clean_county'] = county_and_pop['County'].map(canonicalize_county)
county_and_state['clean_county'] = county_and_state['County'].map(canonicalize_county)
````
